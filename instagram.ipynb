```python
import numpy as np
import matplotlib.pyplot as plt
from random import randrange

# 生成示例数据
x = np.arange(1, 11)  # 生成 x 值，从 1 到 10
y = np.array([randrange(10, 30) for _ in range(10)])  # 生成随机的 y 值

# 绘制折线图
plt.plot(x, y, label="原始数据", marker='o')

# 使用NumPy的polyfit函数拟合趋势线
degree = 1  # 线性趋势
coefficients = np.polyfit(x, y, degree)
trendline = np.poly1d(coefficients)

# 生成趋势线上的点
trendline_x = np.linspace(x.min(), x.max(), 100)
trendline_y = trendline(trendline_x)

# 绘制趋势线
plt.plot(trendline_x, trendline_y, label="趋势线", linestyle='--')

# 添加标签和标题
plt.xlabel("X轴")
plt.ylabel("Y轴")
plt.title("折线图与趋势线")

# 添加图例
plt.legend()

# 显示图形
plt.grid(True)
plt.show()
```
